# **Theory questions:-**

**1. What is a random variable in probability theory?**  
- A random variable is a variable that takes numerical values determined by the outcome of a random phenomenon.
---
**2. What are the types of random variables?**  
- There are two types: discrete random variables (finite values) and continuous random variables (infinite values within a range).
---
**3. What is the difference between discrete and continuous distributions?**  
- Discrete distributions deal with countable outcomes, while continuous distributions deal with uncountable outcomes over an interval.
---
**4. What are probability distribution functions (PDF)?**  
- PDF defines the likelihood of a random variable taking a specific value in a continuous distribution.
---
**5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**  
- CDF gives the probability that a random variable is less than or equal to a certain value, while PDF gives the probability density at a specific value.
---
**6. What is a discrete uniform distribution?**  
- A discrete uniform distribution assigns equal probability to all outcomes in a finite set.
---
**7. What are the key properties of a Bernoulli distribution?**  
- It has two possible outcomes (success or failure) and is defined by a single parameter \(p\) (probability of success).
---
**8. What is the binomial distribution, and how is it used in probability?**  
- The binomial distribution models the number of successes in a fixed number of trials, each with the same probability of success.
---
**9. What is the Poisson distribution and where is it applied?**  
- The Poisson distribution models the probability of a given number of events occurring in a fixed interval of time or space.
---
**10. What is a continuous uniform distribution?**  
- A continuous uniform distribution has constant probability density over a certain interval.
---
**11. What are the characteristics of a normal distribution?**  
- It is symmetric, bell-shaped, and defined by its mean and standard deviation.
---
**12. What is the standard normal distribution, and why is it important?**  
- It is a normal distribution with a mean of 0 and a standard deviation of 1. It simplifies statistical analysis.
---
**13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?**  
- CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population's distribution.
---
**14. How does the Central Limit Theorem relate to the normal distribution?**  
- It explains why many random variables, when summed or averaged, form a normal distribution.
---
**15. What is the application of Z statistics in hypothesis testing?**  
- Z-statistics are used to test hypotheses about population means when the population standard deviation is known.
---
**16. How do you calculate a Z-score, and what does it represent?**  
- \( Z = \frac{(X - \mu)}{\sigma} \), where \(X\) is the data point, \(\mu\) is the mean, and \(\sigma\) is the standard deviation. It represents how many standard deviations a data point is from the mean.
---
**17. What are point estimates and interval estimates in statistics?**  
- Point estimates provide a single value estimate of a parameter, while interval estimates provide a range of values within which the parameter lies.
---
**18. What is the significance of confidence intervals in statistical analysis?**  
- Confidence intervals indicate the range within which a population parameter is likely to fall with a certain confidence level.
---
**19. What is the relationship between a Z-score and a confidence interval?**  
- Z-scores determine the critical values used to calculate confidence intervals.
---
**20. How are Z-scores used to compare different distributions?**  
- Z-scores standardize data from different distributions, allowing comparison on a common scale.
---
**21. What are the assumptions for applying the Central Limit Theorem?**  
- The sample size should be sufficiently large, and the data should be independent and identically distributed.
---
**22. What is the concept of expected value in a probability distribution?**  
- The expected value is the weighted average of all possible outcomes of a random variable.
---
**23. How does a probability distribution relate to the expected outcome of a random variable?**  
- The probability distribution determines the likelihood of each outcome, which is used to calculate the expected value.

---

```python
# 2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)
import matplotlib.pyplot as plt
import numpy as np

# Discrete uniform distribution (example: rolling a 6-sided die)
values = np.arange(1, 7)
probabilities = [1/6] * 6  # Equal probabilities for all sides

# Plotting PMF
plt.stem(values, probabilities)  # Removed `use_line_collection`
plt.title("PMF of Discrete Uniform Distribution (Dice Roll)")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.show()

```

```python
# 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
from scipy.stats import bernoulli

def bernoulli_pdf(p, x):
    return bernoulli.pmf(x, p)

# Example: Bernoulli distribution with p=0.5
p = 0.5
x_values = [0, 1]
pdf_values = [bernoulli_pdf(p, x) for x in x_values]
print("PDF Values:", pdf_values)
```

```python
#4. Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram.

import matplotlib.pyplot as plt
import numpy as np

# Simulate binomial distribution
n, p = 10, 0.5
data = np.random.binomial(n, p, 1000)

# Plot histogram
plt.hist(data, bins=range(0, n+2), density=True, alpha=0.7, color='blue', edgecolor='black')
plt.title("Histogram of Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

```python
#5. Create a Poisson distribution and visualize it using Python.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
lambda_value = 4  # Mean of the Poisson distribution
x = np.arange(0, 15)  # Range of possible events

# PMF of Poisson distribution
pmf = poisson.pmf(x, lambda_value)

# Plotting
plt.stem(x, pmf)  # Removed `use_line_collection`
plt.title("Poisson Distribution (=4)")
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.show()

plt.show()
```

```python
#6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.
import numpy as np
import matplotlib.pyplot as plt

# Discrete uniform distribution
values = np.arange(1, 7)
cdf = np.cumsum([1/6] * 6)  # CDF

# Plotting
plt.step(values, cdf, where="post")
plt.title("CDF of Discrete Uniform Distribution (Dice Roll)")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.show()
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    return (data - mean) / std_dev

# Example dataset
data = np.random.normal(50, 10, 1000)
z_scores = calculate_z_scores(data)

# Plot Z-scores
plt.hist(z_scores, bins=30, density=True, alpha=0.7, color='orange', edgecolor='black')
plt.title("Z-Scores of Dataset")
plt.xlabel("Z-Score")
plt.ylabel("Frequency")
plt.show()
```

```python
#10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import matplotlib.pyplot as plt

# Non-normal distribution (exponential)
population = np.random.exponential(scale=2, size=100000)

# Central Limit Theorem: Means of samples
sample_means = [np.mean(np.random.choice(population, size=30)) for _ in range(1000)]

# Plot histogram
plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='teal', edgecolor='black')
plt.title("Central Limit Theorem Demonstration")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.show()
```

```python
#16. Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1).
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

def standard_normal_distribution():
    x = np.linspace(-4, 4, 100)
    pdf = norm.pdf(x, 0, 1)
    plt.plot(x, pdf, label="Standard Normal Distribution")
    plt.title("Standard Normal Distribution (Mean=0, Std=1)")
    plt.xlabel("x")
    plt.ylabel("Density")
    plt.legend()
    plt.show()

standard_normal_distribution()
```

```python
#17. Generate random variables and calculate their corresponding probabilities using the binomial distribution.
from scipy.stats import binom

# Parameters
n, p = 10, 0.5
x_values = range(n + 1)
probabilities = binom.pmf(x_values, n, p)

# Display probabilities
for x, prob in zip(x_values, probabilities):
    print(f"P(X={x}) = {prob:.4f}")
```

```python
#18. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
from scipy.stats import norm

def calculate_z_score(value, mean, std_dev):
    return (value - mean) / std_dev

# Example
value, mean, std_dev = 75, 70, 5
z_score = calculate_z_score(value, mean, std_dev)
probability = norm.cdf(z_score)

print(f"Z-Score: {z_score}")
print(f"Probability of being below this value: {probability:.4f}")
```

```python
#19. Implement hypothesis testing using Z-statistics for a sample dataset.
from scipy.stats import norm

# Example dataset
sample_mean = 75
population_mean = 70
std_dev = 5
sample_size = 30

# Z-statistic
z_stat = (sample_mean - population_mean) / (std_dev / np.sqrt(sample_size))
p_value = 1 - norm.cdf(z_stat)

print(f"Z-Statistic: {z_stat}")
print(f"P-Value: {p_value:.4f}")
```

```python
#20. Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
from scipy.stats import norm

# Example dataset
data = np.random.normal(50, 10, 100)
mean = np.mean(data)
std_error = np.std(data) / np.sqrt(len(data))
confidence = 0.95
z_value = norm.ppf((1 + confidence) / 2)

# Confidence interval
lower = mean - z_value * std_error
upper = mean + z_value * std_error
print(f"95% Confidence Interval: [{lower:.2f}, {upper:.2f}]")
```

```python
#21. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
# Normal distribution data
data = np.random.normal(60, 15, 200)

# Confidence interval (reusing above logic)
mean = np.mean(data)
std_error = np.std(data) / np.sqrt(len(data))
confidence = 0.95
z_value = norm.ppf((1 + confidence) / 2)

lower = mean - z_value * std_error
upper = mean + z_value * std_error
print(f"95% Confidence Interval for the mean: [{lower:.2f}, {upper:.2f}]")
```

```python
#22. Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# Normal distribution parameters
mu, sigma = 50, 10
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
pdf = norm.pdf(x, mu, sigma)

# Plotting
plt.plot(x, pdf, label="PDF")
plt.title("Probability Density Function (Normal Distribution)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```
